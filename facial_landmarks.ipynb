```python
import dlib
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

%matplotlib inline
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('./models/shape_predictor_68_face_landmarks.dat')
image = cv2.imread('./images/AH.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
bounds = detector(image, 1)[0]
```

```python
fig, ax = plt.subplots(1)
ax.imshow(image, cmap='Greys_r')
patch = patches.Rectangle((bounds.left(),bounds.top()),bounds.right() - bounds.left(), bounds.bottom() - bounds.top(),linewidth=1,edgecolor='r',facecolor='none')
ax.add_patch(patch)
plt.show()
```

```python
landmarks = predictor(image, bounds)
points = list(landmarks.parts())
fig, ax = plt.subplots(1)
ax.imshow(image, cmap='Greys_r')
for p in points:
    point = patches.CirclePolygon((p.x, p.y), radius=10, color='blue')
    ax.add_patch(point)
    
plt.show()
```
